## Linear Regression

```python
"""
This is colorcoding each team by color, plotting against the wins, givin a specific league.
"""
Y = "xGA"
X = "position"
League = "EPL"

df = data.copy()
df = df.loc[df["league"] == League]
# plt.scatter(df["wisn"],df["team"], color='red')
for i in df["team"].unique():
    x = df.loc[df["team"] == i]
    plt.scatter(x[X],x[Y])
    # print(x["team"].uniq.unique()ue(), x["wins"].std())

firstP = df.loc[df[X].idxmin(), [X, Y]]#[Y]
lastP = df.loc[df[X].idxmax(), [X, Y]]#[Y]
# print(firstP, lastP)

if firstP[Y] > lastP[Y]:
    # x = np.arange(min(df[X]),max(df[X]) + 1, 1)
    y = (max(df[Y]) - min(df[Y]))/(min(df[X]) - max(df[X]))* (df[X] - min(df[X])) + max(df[Y]) 
    plt.plot(df[X],y)
    # print(max(df[Y]),  min(df[Y]))
    # print(max(df[X]),  min(df[X]))
else:
    # x = np.arange(min(df[X]),max(df[X]) + 1, 1)
    y = -(max(df[Y]) - min(df[Y]))/(min(df[X]) - max(df[X]))* (df[X] - min(df[X])) + min(df[Y])
    plt.plot(df[X],y)


def mse(yTrue, yPred):
    yTrue, yPred = list(yTrue), list(yPred)
    total = 0
    #Ypred is a list
    for i in range(0, len(yPred)):
        total += (yTrue[i]-yPred[i]) ** 2
    total /= len(yPred)
    return total

# Analytical linear regression slope
def linearRegression(x, y):
    x,y = np.array(x), np.array(y)
    x_mean, y_mean = np.mean(x), np.mean(y)
    a, b = 0, 0
    for i in range(0, len(x)):
        a += ((x[i] - x_mean) * (y[i] - y_mean))
        b += ((x[i] - x_mean) ** 2)
    
    slope = a/b
    y_int = y_mean - slope * x_mean

    y = slope * x + y_int
    return x, y

x_new, y_new = (linearRegression(df[X], df[Y]))
plt.plot(x_new,y_new)

a = np.ones([len(df[X])])
a *= np.mean(df[Y])
plt.plot(df[X], a)



# print(y, len(y))
# print(list(df[Y]))

#Printing results

# drops all the indecies to 0

y.reset_index(inplace = True,drop = True)

print("My line ", mse(y, df[Y]))
# r2 value for my line
print("My line r2", r2_score(y, list(df[Y])))


print("Mean ", mse(df[Y], a))
# r2 value for the mean
print("mean r2", r2_score(list(df[Y]), list(a)))

print("Equation ", mse(y_new, df[Y]))

# r2 value for the equation
print("Equation r2", r2_score(list(df[Y]), list(y_new)))
```

```python
# iterative linear regression
# x = np.arange(1, 6, 1)
# print(x[::-1])

#new_df_x = df[X][::-1]

new_df_x = np.array(df[X])

plt.figure()
plt.scatter(new_df_x, df[Y])
lr = 0.002
m, b = 0.5,15
n = 100 # number of iterations


graph_vars = np.zeros((n,3))

def step_grab(x,y, m, b, lr):
    y = np.array(y)
    m_grad, b_grad = 0, 0
    
    for i in range(len(x)):
        x0 = x[i]
        y0 = y[i]
        m_grad += -(2/len(x)) * x0 * (y0 - (m * x0 + b))
        b_grad += -(2/len(x)) * (y0 - (m * x0 + b))
    
    m -= m_grad/2 * lr
    b -= b_grad * lr
    return m, b

for i in range(n):

    plt.plot(new_df_x, m*new_df_x + b)
    m, b = step_grab(new_df_x, df[Y], m, b, lr)
    #print(m, b, mse(df[Y], m*new_df_x + b))
    graph_vars[i,:] = [m, b, mse(df[Y], m*new_df_x + b)]



#print(graph_vars)

```

```python
# Graph loss surface

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)



n = 20
slopes = np.linspace(0, 3, n)
y_ints = np.linspace(0, 25, n)
loss_surface = np.zeros([n**2,3])


# Putting slope into the array
for i in range(n):

    for j in range(n):
        loss_surface[i * n + j, 0] = slopes[i]
    
        #print(y_int[i])
        loss_surface[i * n + j, 1] = y_ints[j]
        

        loss_surface[i * n + j, 2] = mse(df[Y], slopes[i] * new_df_x + y_ints[j])
    



ax.scatter(loss_surface[:, 0], loss_surface[:, 1], loss_surface[:, 2])
ax.scatter(graph_vars[:,0], graph_vars[:,1], graph_vars[:,2], s= 50)

print(loss_surface[:5, :])

```

```python
trainsplit = 0.8
y_col = ["deep"]
x_cols = [i for i in num_data.columns if i not in y_col]

x_train, y_train = num_data.loc[:len(num_data) * trainsplit,  x_cols], num_data.loc[:len(num_data) * trainsplit,  y_col]
x_valid, y_valid = num_data.loc[len(num_data) * trainsplit:,  x_cols], num_data.loc[len(num_data) * trainsplit:,  y_col]
print(len(x_train), len(y_train), len(x_valid), len(y_valid))

```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
# start the model
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
print("\ndecision tree")
print(model.get_depth())
print(model.score(x_train, y_train))
print(model.score(x_valid, y_valid))

model = RandomForestRegressor()
model.fit(x_train, y_train)
print("\nrandom forest")
print(model.score(x_train, y_train))
print(model.score(x_valid, y_valid))

```
